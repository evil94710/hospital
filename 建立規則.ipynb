```python
class Apriori():
    
    def __init__(self, dis, result):
        self.dis = dis
        self.result = result
    
    def apriori_result(self,dis_count):
        print(dis_count, dis[dis_count])
        string1 = 'C:\\Users\\user\\Downloads\\聖馬爾定\\健保資料完整性\\疾病號\\' + dis[dis_count] + '.csv'
        df = pd.read_csv(string1)
        df.drop(columns='Unnamed: 0',inplace = True)
        frequent_itemsets = apriori(df, min_support=0.3,use_colnames=True,verbose=1,max_len=6)
        frequent_itemsets.sort_values("support", ascending=False,inplace=True)
        itemset = frequent_itemsets['itemsets'].values.tolist()
        print('apriori finish')
        list_B = []
        for i in itemset:
            temp = []
            for j in i:
                r = j
                temp.append(r)
            list_B.append(temp)
        for i in list_B:
            i.sort()
        temp = []
        for i in range(len(list_B)):
            if len(temp) < 50:
                temp.append(list_B[i])
            else:
                break
        result[dis[dis_count]] = temp
        del df,frequent_itemsets,itemset
        gc.collect()
```

```python
Apriori_model = Apriori(dis,result)
```

```python
for i in range(len(dis)):
    Apriori_model.apriori_result(i)
# for i in range(len(dis)):
#     Apriori_model.apriori_result(i)
```
